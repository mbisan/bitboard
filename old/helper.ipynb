```python
def diagIndices(i, j, k, h):
    res = []
    if i<k and j<h:
        s, t = i, j
        while (s!=k and t!=h):
            res.append((s, t))
            s+=1
            t+=1
        res.append((s, t))
        return res
    if i>k and j>h:
        s, t = k, h
        while (s!=i and t!=j):
            res.append((s, t))
            s+=1
            t+=1
        res.append((s, t))
        return res
    return []
```

```python
lookup = np.zeros((8, 8, 8, 8), dtype=np.uint64)
for i in range(8):
    for j in range(8):
        for k in range(8):
            for h in range(8):
                lookup[i, j, k, h] += 2**(8*i+j)
                lookup[i, j, k, h] += 2**(8*k+h)
                if (i == k): # same file
                    if j<h:
                        for a in range(j, h):
                            lookup[i, j, k, h] += 2**(8*i+a)
                    else:
                        for a in range(h, j):
                            lookup[i, j, k, h] += 2**(8*i+a)
                if (j == h): # same col
                    if i<k:
                        for a in range(i, k):
                            lookup[i, j, k, h] += 2**(8*a+j)
                    else:
                        for a in range(k, i):
                            lookup[i, j, k, h] += 2**(8*a+j)
                if ((i+j) == (k+h)): # same diag
                    if i<j:
                        for a in range(min(i, k), max(i, k), 1):
                            lookup[i, j, k, h] += 2**(8*(a)+(i+j-a))
                    else:
                        for a in range(min(j, h), max(j, h), 1):
                            lookup[i, j, k, h] += 2**(8*(i+j-a)+(a))
                if ((-i+j) == (-k+h)): # same diag
                    for s, t in diagIndices(i, j, k, h):
                        lookup[i, j, k, h] += 2**(8*(s)+(t))
```
