```python
def processindex(i, j, k, l):
    possibleIndices = []
    if i==k and j==l:
        return 0
    if i==k: # same row
        if j<l:
            for t in range(8):
                possibleIndices.append((i, l-t))
        else:
            for t in range(8):
                possibleIndices.append((i, l+t))
        for t in range(8):
            possibleIndices.append((t, j))
    if j==l: # same col
        if i<k:
            for t in range(8):
                possibleIndices.append((k-t, j))
        else:
            for t in range(8):
                possibleIndices.append((k+t, j))
        for t in range(8):
            possibleIndices.append((i, t))
    
    if (i+j) == (k+l) or (i-j) == (k-l): # same diag
        if i<k and j<l:
            for t in range(8):
                possibleIndices.append((k-t, l-t))
            for t in range(-5,5,1):
                possibleIndices.append((i+t, j-t))
        if i<k and j>l:
            for t in range(8):
                possibleIndices.append((k-t, l+t))
            for t in range(-5,5,1):
                possibleIndices.append((i+t, j+t))
        if i>k and j<l:
            for t in range(8):
                possibleIndices.append((k+t, l-t))
            for t in range(-5,5,1):
                possibleIndices.append((i+t, j+t))
        if i>k and j>l:
            for t in range(8):
                possibleIndices.append((k+t, l+t))
            for t in range(-5,5,1):
                possibleIndices.append((i+t, j-t))

    actualIndices: np.uint64 = 0
    indicesList = []
    for m, n in list(set(possibleIndices)):
        if m>=0 and m<8 and n>=0 and n<8:
            actualIndices ^= 1<<(8*m+n)
            indicesList.append((m, n))

    return actualIndices
```

```python
lookup=np.zeros((8, 8, 8, 8), dtype=np.uint64)

for i in range(8):
    for j in range(8):
        print("{", end="")
        for k in range(8):
            for l in range(8):
                try:
                    lookup[i, j, k, l] = processindex(i, j, k, l)
                    print(hex(lookup[i, j, k, l]) + "ULL, ", end="")
                except:
                    print(i, j, k, l, processindex(i, j, k, l))
        print("},")
```
