```python
bitpositions = []
for i in range(8):
    for j in range(8):
        bitpositions.append([])
        for k in range(8):
            bitpositions[-1].append((i, k))
        for k in range(8):
            bitpositions[-1].append((k, j))

bitpositions_relevant = []

for i in range(8):
    for j in range(8):
        bitpositions_relevant.append([])

        for bp in bitpositions[i*8+j]:
            if abs(bp[0]-i)>=1 or abs(bp[1]-j)>=1:
                bitpositions_relevant[-1].append(bp[0]*8+bp[1])

        bitpositions_relevant[-1].sort()
```

```python
for p in bitpositions_relevant:
    print(p)
```
